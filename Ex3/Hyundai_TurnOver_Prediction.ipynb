# Hyundai Turnover Prediction

Dự đoán doanh thu của hãng xe Huyndai dòng Elantra trong năm 2013 và đầu 2014 dựa vào dự liệu trước đó
* Month, Year
* ElantraSales (prediction)
* Unemployment
* Queries
* CPI_energy
* CPI_all

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('elantra.csv')
```

```python
df = df.sort_values(by = ['Year', 'Month'])
df = df.reset_index(drop = True)
```

```python
df.head(10)
```

```python
df.tail(10)
```

```python
plt.figure(figsize = (9,6))
plt.plot(df.ElantraSales.values)
plt.xlabel("Time index")
plt.ylabel("Sales")
plt.show()
```

```python
numeric_features = df.columns.drop(['ElantraSales', 'Month', 'Year'])
numeric_features
```

```python
df_train = df[df.Year < 2013]
df_test = df[df.Year >= 2013]

y_train = df_train.ElantraSales.values
y_test = df_test.ElantraSales.values
```

```python
# Features scaling
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler().fit(df_train[numeric_features]) # compute mean, std

X_train = scaler.transform(df_train[numeric_features]) # scale value based on mean, std which learned on fit() step
X_test = scaler.transform(df_test[numeric_features])
```

## Build Model

```python
from sklearn.linear_model import LinearRegression

lregr = LinearRegression()
lregr.fit(X_train, y_train)
```

## Evaluation

```python
from sklearn.metrics import mean_squared_error

def relative_error(y_true, y_pred):
    errors = np.abs(y_pred - y_true).astype(float) / y_true
    return np.mean(errors) * 100
```

```python
y_pred_test = lregr.predict(X_test)
print('RMSE: {:.2f}'.format(np.sqrt(mean_squared_error(y_test, y_pred_test))))
print('Relative error: {:.1f}%'.format(relative_error(y_test, y_pred_test)))
```

```python
# Compare y_test and y_pred_test
plt.figure(figsize = (9,6))
plt.plot(y_test)
plt.plot(y_pred_test)
plt.xlabel('Time index')
plt.ylabel('Sales')
```

Có thể nhận thấy kết quả dự đoán của model không khớp với dữ liệu thực tế
→ Cần cải thiện model, tận dụng feature 'Month' hiệu quả hơn

## Improvement

```python
month_onehot_train = pd.get_dummies(df_train.Month) # convert from categorical into numeric (one-hot encoding)
month_onehot_train
```

```python
X_train = np.hstack((X_train, month_onehot_train))
```

```python
X_test = np.hstack((X_test, pd.get_dummies(df_test.Month)))
```

```python
lregr.fit(X_train, y_train)
```

```python
# Evaluation
y_pred_test = lregr.predict(X_test)
print('RMSE: {:.2f}'.format(np.sqrt(mean_squared_error(y_test, y_pred_test))))
print('Relative error: {:.1f}%'.format(relative_error(y_test, y_pred_test)))
```

```python
# Compare y_test and y_pred_test
plt.figure(figsize = (9,6))
plt.plot(y_test)
plt.plot(y_pred_test)
plt.xlabel('Time index')
plt.ylabel('Sales')
```
